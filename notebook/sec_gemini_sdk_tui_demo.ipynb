## Authentication

Get your API key and use it to authenticate.


````python
import os

# Get the API key from environment variables.
API_KEY = os.environ.get("SEC_GEMINI_API_KEY", "")
````

## Connecting to the Service

Connect to the Sec-Gemini service using the SDK.


````python
from sec_gemini import SecGemini

client = SecGemini(api_key=API_KEY)
await client.start()
print("Connected to Sec-Gemini")
````

## Creating a Session

Create a session to interact with the service.


````python
session = await client.sessions.create()
print(f"Session Created: {session.id}")
````

## Using MCPs

List and use MCPs in the session.


````python
skills_mcps = await client.skills.list()
for mcp in skills_mcps:
    for skill in mcp.skills:
        print(f"Skill: {skill}")
````

## Submitting a Query

Submit a prompt to the session.


````python
await session.prompt("What is the capital of France?")

async for msg in session.messages.stream():
    msg_type = msg.get("message_type", "")
    content = msg.get("content", "")

    if msg_type == "MESSAGE_TYPE_RESPONSE":
        print(f"Response: {content}")
    elif msg_type == "MESSAGE_TYPE_AGENT_IS_DONE":
        break
````

## Using the TUI

The TUI provides an interactive terminal interface to manage sessions.
You can launch it from the command line:

```bash
pip install sec-gemini
sec-gemini
```

You can also list sessions programmatically:

````python
sessions = await client.sessions.list()
for s in sessions:
    print(f"Session: {s.id}  Name: {s.name}  Status: {s.status}")
````

## Cleanup

````python
await session.delete()
await client.close()
print("Done!")
````
